```python
import pandas as pd
```

```python
skupna = pd.read_csv('podatki/nepremicnine.csv')

del skupna['posred']
skupna.columns = ['ime', 'vrsta', 'opis', 'leto_izgradnje', 'zemljisce', 'velikost', 'cena', 'agencija', 'regija']
```

```python
skupna.insert(0, 'id', range(1, 1 + len(skupna)))
skupna
```

```python
nepremicnine = skupna
nepremicnine = nepremicnine.drop('agencija', 1)
nepremicnine = nepremicnine.drop('regija', 1)
nepremicnine.to_csv('tabele/nepremicnine.csv', index=False)
nepremicnine
```

```python
data = ['ljubljana-mesto', 'ljubljana-okolica', 'gorenjska', 'juzna-primorska', 'severna-primorska',
            'notranjska', 'savinjska', 'podravska', 'koroska', 'dolenjska', 'posavska', 'zasavska', 'pomurska']
regije = pd.DataFrame(data, columns = ['regije'])
regije.insert(0, 'id', range(1, 1 + len(regije)))
regije.to_csv('tabele/regije.csv', index=False)
regije
```

```python
data = skupna.agencija.unique()
agencije = pd.DataFrame(data, columns = ['agencija'])
agencije.insert(0, 'id', range(1, 1 + len(agencije)))
agencije.to_csv('tabele/agencije.csv', index=False)
agencije
```
